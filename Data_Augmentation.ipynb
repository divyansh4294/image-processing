In this notebook, we are going to see how we can do data augmentation on an image of cat and dog.
1. Data augmentation helps us to increase the dataset size. Suppose we have 500 images of cat but to train a cnn and making it robust then this much data is not sufficient. So we need data augmentation.

2. Data Augmentation introduces invariance to the dataset.



We are implemention data augmentation using keras image data generator


```
"""
@Author = Divyansh.Gupta
"""
# Importing functions from keras preprocessing 
# 1. ImageDataGenerator: We will use this to generate images
# 2. arry_to_img: used to convert array into images
# 3. img_to_array: used to convert images into array
# 4. load_img: used to read the image.
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```
# This is the sample cat image
img = load_img("cat1.jpg")
img
```

```
x = img_to_array(img)
print("shape of image:",x.shape)
x = x.reshape((1,)+x.shape)
print("Reshape image size:",x.shape)
```

# Rotation

```
# Changing rotation range to 40 and keeping all parameter as default
rotation = ImageDataGenerator(
    featurewise_center=False, 
    samplewise_center=False, 
    featurewise_std_normalization=False, 
    samplewise_std_normalization=False, 
    zca_whitening=False, 
    zca_epsilon=1e-06, 
    rotation_range=40, # Int. Degree range for random rotations
    width_shift_range=0.0, 
    height_shift_range=0.0, 
    brightness_range=None, 
    shear_range=0.0, 
    zoom_range=0.0, 
    channel_shift_range=0.0, 
    fill_mode='nearest', 
    cval=0.0, 
    horizontal_flip=False, 
    vertical_flip=False, 
    rescale=None, 
    preprocessing_function=None, 
    data_format=None, 
    validation_split=0.0, 
    dtype=None
    
)
```

```
i=0
for batch in rotation.flow(x,
                           batch_size=1,
                           save_to_dir="sample_data",
                           save_prefix="cat",
                           save_format="jpeg"
                           ):
  
  if i == 0:
    break
  
```

```
# after rotation
print("Before Rotation")
img
```

```
print("After Rotation")
rot_image = load_img("sample_data/cat_0_1032.jpeg")
rot_image
```

# Applying all together and generate data 

```
dataset = ImageDataGenerator(
    featurewise_center=False, 
    samplewise_center=False, 
    featurewise_std_normalization=False, 
    samplewise_std_normalization=False, 
    zca_whitening=False, 
    zca_epsilon=1e-06, 
    rotation_range=40, # Change
    width_shift_range=0.2, # Change
    height_shift_range=0.2, # Change
    brightness_range=None, 
    shear_range=0.2, # Change
    zoom_range=0.2, # Change
    channel_shift_range=0.0,
    fill_mode='nearest', 
    cval=0.0, 
    horizontal_flip=True, # If true: it'll flip the image horizontally 
    vertical_flip=False, # If true: it'll flip the image vertically
    rescale=None, 
    preprocessing_function=None, 
    data_format=None, 
    validation_split=0.0, 
    dtype=None
    
)
```

```
i=0
n = 3 # number of images augmented using image
for batch in dataset.flow(x,
                           batch_size=1, # this much image would be consider at once
                           save_to_dir="sample_data", # Data get save into folder
                           save_prefix="cat", # image name have prefix
                           save_format="jpeg" # image have this format
                           ):
  
  if i > n:  
    break
```
